## 一、收集数据
收集豆瓣读书“日系推理”tag前50页书籍的书名、评分等信息。

```python
import pandas as pd
import os
import requests
from bs4 import BeautifulSoup
```

```python
info_urls=[]
for i in range(79):
    num = i*20
    info_urls.append('https://book.douban.com/tag/日系推理?start='+str(num)+'&type=T')
```

```python
i = 0
for url in info_urls:
    if (i<50):
        r = requests.get(url)
        txt = r.text
        with open('C:\\Users\\Opsuc\\learning\\project\\pages\\'+str(i*20)+'.html','w',encoding='utf-8') as f:  
            f.write(txt)
        i = i+1
```

```python
all_list=[]
folder = 'pages'
for html in os.listdir(folder):
    with open(os.path.join(folder,html),'rb') as file:
        soup = BeautifulSoup(file)
    num = 0
    links = []
    for link in soup.find_all('a',class_ ='nbg'):
        links.append(link.get('href'))
        
    for i in range(20):   
        rate_num = soup.find_all('span',class_ = "pl")[i].contents[0][len('\r\n        ('):-len('人评价)\r\n    ')]
        if (rate_num=='少于10' or rate_num=='目前无'):
            num = num + 1
            score = '0'
        elif(rate_num=='10'or rate_num=='11'or rate_num=='12'):
            score = '0'        
        else:
            score = soup.find_all('span',class_ = "rating_nums")[(i - num)].contents[0]

                    
        info = soup.find_all('div',class_ = "pub")[i].contents[0][len('\r\n        \r\n  \r\n  '):-len('\r\n\r\n      ')]
        title = soup.find_all(title=True)[i].contents[0].split('\r\n')[2].strip()
        if(len(info.split('/'))>=4):
            writer = info.split('/')[0]
            press = info.split('/')[2].strip()
            date = info.split('/')[3].strip()
        else:
            writer = info.split('/')[0]
            press = '无'
            date = '无'
        link = links[i]
        all_list.append({'title':title,
                        'writer':writer,
                       'score':score,
                       'rate_num':rate_num,
                        'press':press,
                        'date':date,
                         'link':link
                       })
dfall = pd.DataFrame(all_list, columns =['title','writer','rate_num','score','press','date','link'])
dfall

    
```

## 二、清洗数据
#### 目前数据存在的问题有：
- 有的书具有多个版本，可能需要汇总。
- 一些书的作者、出版社信息和出版日期信息缺失。
- 作者名不唯一。有些标注了国籍，有些没有。大部分是译名，但也有原名。还有繁体、简体的问题。


```python
import re
dfall_clean = dfall.copy()
dfall_clean['writer'] = dfall_clean['writer'].str.replace(' ','')
writer = dfall_clean.writer[dfall_clean['writer'].str.contains('[日]')].str[3:]
for index in writer.index:
    dfall_clean['writer'].loc[index] = writer.loc[index]
dfall_clean
```

#### 将作者名进行了统一

```python
without_press = dfall_clean[dfall_clean['press'].str.contains('无')]
num_without_rate = len(dfall_clean[dfall_clean['score'].str.startswith('0')])
num_with_wrong_press_info=len(dfall_clean[dfall_clean['press'].str.contains('[0-9]',regex = True)])
with_wrong_press = dfall_clean[dfall_clean['press'].str.contains('[0-9]',regex = True)]
with_wrong_date = dfall_clean[dfall_clean['date'].str.contains(r'^\D',regex = True)]
wrong_info = without_press.append(with_wrong_press).append(with_wrong_date)
wrong_info =wrong_info.drop_duplicates()
wrong_info
```

#### 以上是数据有误的行

```python
dfall_clean_1 = dfall_clean.copy()
import re
for index in wrong_info.index:
    r = requests.get(wrong_info.loc[index]['link'])
    txt = r.text
    with open('C:\\Users\\Opsuc\\learning\\project\\wrongpages\\'+str(index)+'.html','w',encoding='utf-8') as f:  
        f.write(txt)
    with open('wrongpages\\'+str(index)+'.html','rb') as file:
        soup = BeautifulSoup(file)
    for item in str(soup.find_all('div',id ='info')).split('<br/>'):
        if ('出版社' in item):
            press = item[len('''\n<span class="pl">出版社:</span> '''):]
        if('出版年'in item):
            date = item[len('''\n<span class="pl">出版年:</span> '''):]
        else:
            date = '2019-1-1'
    dfall_clean['press'].loc[index] = press
    dfall_clean['date'].loc[index] = date
dfall_clean
```

#### 通过对书的详情页面信息的再次抓取，补充出版社、出版日期信息

```python
for index in range(1000):
    item = dfall_clean['date'].iloc[index]
    
    if '.' in item:
        split_date = item.split('.')
    elif '-' in item:
        split_date = item.split('-')
    elif '/' in item:
        split_date = item.split('/')
    elif '年' in item:
        regex = re.compile(r'(\d+)年((\d+)月)?((\d+)日)?')
        mo = regex.search(item)
        if '日' in mo.group():
            split_date=[mo.group(1),mo.group(3),mo.group(5)]
        elif '月' in mo.group():
            split_date =[mo.group(1),mo.group(3)]
        else:
            split_date =[mo.group(1)]
    else:
        split_date = [item]
    
    if (len(split_date) == 1):
        date = str(split_date[0])+'-1'+'-1'
    elif (len(split_date) == 2):
        date = '-'.join(split_date)+'-1'
    else:
        date = '-'.join(split_date)
    
    dfall_clean['date'].iloc[index] = date
dfall_clean
```

#### 将日期的格式统一后，大致的数据清理到这里就完成了
